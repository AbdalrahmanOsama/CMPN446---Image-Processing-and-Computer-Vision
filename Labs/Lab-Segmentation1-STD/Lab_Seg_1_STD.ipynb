```python
# 1- Get the required RGB color (Using a color picker)


# 2- Read image

# 3- extract R, G and B channels (as float)

# 4- calculate differences FOR EACH CHANNEL (between the image and the required pixel value)

# 5- calculate overall distance from the given RGB color (use any appropriate distance measure, take care of the minus distances)

# 6-  create a mask by thresholding the differences

# 7- In the input image, Replace the pixels of the mask with the following color
# R = 230 , G = 90 , B=40

# 8- show the image or save it


image2 = io.imread('imgs/exp2/cufe.png')
img2Copy=image2.copy()
image2_float = image2.astype(float)

red_image2 = image2_float[:,:,0]
green_image2 = image2_float[:,:,1]
blue_image2 = image2_float[:,:,2]
show_images([red_image2,green_image2,blue_image2],["Red image","Green image","Blue image"])

red_image2_diff = np.abs(red_image2 - 84)
green_image2_diff = np.abs(green_image2 - 95)
blue_image2_diff = np.abs(blue_image2 - 41)

overall_distance = np.sqrt(np.power(red_image2_diff,2)+np.power(green_image2_diff,2)+np.power(blue_image2_diff,2))

threshold = overall_distance < 29
show_images([threshold])

img2Copy[:,:,0][threshold==1]=230
img2Copy[:,:,1][threshold==1]=90
img2Copy[:,:,2][threshold==1]=40

show_images([img2Copy])
```
